```python
uri = "https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/master/aula4.1/ratings.csv"
notas = pd.read_csv(uri)
notas.head()
```

```python
notas.columns = ["UsuárioId", "filmeId", "nota", "Momento"]
notas.head()
```

```python
notas["nota"] #serie de números
```

```python
notas["nota"].unique() #somente os únicos
```

```python
notas["nota"].mean() # exibe a média
```

```python
notas["nota"].min() # exibe nota mínima
```

```python
notas["nota"].max() # exibe nota míáxima
```

```python
notas.describe() #descrição do dataframe
#cuidado ao usar média, pois poderá enganar. 50% corresponde à média.
```

```python
notas.nota.plot(kind="hist")
```
