```python
def calcular_palavra(palavra):
  palavra=input('Digite uma palavra:')
  if len(palavra) < 10:
    print ('Tente uma palavra maior')
  elif 11 <= len(palavra) <=20:
    print('Boa palavra')
  elif len(palavra)>=21:
    print('Uau!')
  else:
    print('Palavra com 10 caracteres')
    
calcular_palavra('eu')
calcular_palavra('documentação')

#coloquei nesse else as situações onde a palavra tem 10 caracteres. Já que abaixo de 10 não entra as de 10 caracteres
```
